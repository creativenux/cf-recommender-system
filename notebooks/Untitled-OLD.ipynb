```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def preprocess_text(text):
    return ' '.join(str(text).lower().split())
```

```python
books_df['text'] = books_df['book_title'] +' '+ books_df['book_author'] +' '+ books_df['classification_level_1'] +' '+ books_df['classification_level_2'] +' '+ books_df['classification_level_3']
books_df['text'] = books_df['text'].apply(preprocess_text)
```

```python
societies_df['text'] = societies_df['society_name'] +' '+ societies_df['society_type'] +' '+ societies_df['society_description'] +' '+ societies_df['society_keywords']
societies_df['text'] = societies_df['text'].apply(preprocess_text)
```

```python
sports_df['text'] = sports_df['sport_name'] +' '+ sports_df['sport_description'] +' '+ sports_df['sport_keywords']
sports_df['text'] = sports_df['text'].apply(preprocess_text)
```

```python
volunteer_programs_df['text'] = volunteer_programs_df['program_name'] +' '+ volunteer_programs_df['program_description'] +' '+ volunteer_programs_df['program_keywords']
volunteer_programs_df['text'] = volunteer_programs_df['text'].apply(preprocess_text)
```

```python
tfidf = TfidfVectorizer(stop_words='english')

books_tfidf = tfidf.fit_transform(books_df['text'])
societies_tfidf = tfidf.fit_transform(societies_df['text'])
sports_tfidf = tfidf.fit_transform(sports_df['text'])
volunteer_programs_tfidf = tfidf.fit_transform(volunteer_programs_df['text'])
```

```python
def get_user_input():
    print("Collaborative Filtering Recommender System")
    
    countries = country_sports_df['country'].unique().tolist()
    print("Please select your country")
    
    for i, country in enumerate(countries, 1):
        print(f"{i}. {country}")
        
    country_index = int(input("Enter the number of your country")) - 1
    country = countries[country_index]
    
    print("Select your top 3 academic interest (enter number separated by column)")
    for i, interest in enumerate(academic_interests, 1):
        print(f"{i}. {interest}")
    
    selected_interest_index = [int(i) - 1 for i in input("Enter your choices").split(',')]
    selected_interest = [academic_interests[i] for i in selected_interest_index]
    
    extra_activities = ['Sports', 'Music', 'Dance', 'Theatre', 'Art', 'Technology', 'Community Service']
    print("Select your top 3 extra activities interest (enter number separated by column)")
    for i, extra_activity in enumerate(extra_activities, 1):
        print(f"{i}. {extra_activity}")
    
    selected_activity_index = [int(i) - 1 for i in input("Enter your choices").split(',')]
    selected_activity = [extra_activities[i] for i in selected_activity_index]
    
    return country, selected_interest, selected_activity
```

```python
def get_recommendations(country, selected_interests, selected_activities, num_recommendation=3):
    print(country, selected_interests, selected_activities)
    
    recommendations = {
        'books': [],
        'societies': [],
        'sports': [],
        'volunteer_programs': []
    }
    
    # get country preferred sport
    preferred_sports = country_sports_df[country_sports_df['country'] == country]['preferred_sport'].values
    if len(preferred_sports) > 0:
        preferred_sports = preferred_sports[0].split(', ')
    else:
        preferred_sports = []
        
    # create profile
    user_profile = ' '.join(selected_interests + selected_activities + preferred_sports)
    user_vector = tfidf.transform([user_profile])
    print(user_vector)
    
    # book recommendation
    books_similarity = cosine_similarity(user_vector, books_tfidf).flatten()
    book_indices = books_similarity.argsort()[::-1][:num_recommendations]
    recommendations['books'] = books_df.iloc[book_indices]['book_title'].tolist()
    
    # Society recommendations
    society_similarity = cosine_similarity(user_vector, societies_tfidf).flatten()
    society_indices = society_similarity.argsort()[::-1][:num_recommendations]
    recommendations['societies'] = societies_df.iloc[society_indices]['society_name'].tolist()
    
    # Sports recommendations
    if preferred_sports:
        sports_mask = sports_df['sport_name'].isin(preferred_sports)
        recommendations['sports'] = sports_df[sports_mask]['sport_name'].tolist()[:num_recommendations]
    else:
        sport_similarity = cosine_similarity(user_vector, sports_tfidf).flatten()
        sport_indices = sport_similarity.argsort()[::-1][:num_recommendations]
        recommendations['sports'] = sports_df.iloc[sport_indices]['sport_name'].tolist()
    
    # Volunteer programs recommendations
    volunteer_similarity = cosine_similarity(user_vector, volunteer_programs_tfidf).flatten()
    volunteer_indices = volunteer_similarity.argsort()[::-1][:num_recommendations]
    recommendations['volunteer_programs'] = volunteer_programs_df.iloc[volunteer_indices]['program_name'].tolist()
    
    return recommendations
    
```

```python
country, interests, extracurriculars = get_user_input()
recommendations = get_recommendations(country, interests, extracurriculars)

print(f"\nRecommendations for a student from {country}:")
for category, items in recommendations.items():
    print(f"\n{category.capitalize()}:")
    for item in items:
        print(f"- {item}")
```
